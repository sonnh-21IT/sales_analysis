* Question 1: Liệu có mối liên hệ nào giữa các thể loại nhạc (genre) và độ phổ biến của bài hát không?

```python
average_popularity_by_genre = data_merged.groupby('genre')['track_popularity'].mean()
# Plotting the Doughnut Chart
plt.figure(figsize=(5, 6))
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(average_popularity_by_genre, labels=average_popularity_by_genre.index, colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Average Track Popularity by Genre')
plt.axis('equal')
plt.show()
```

#### Observations:

1. Sự ưa chuộng của nhạc Pop và Rap: Nhạc Pop và Rap có xu hướng được ưa chuộng hơn các thể loại khác, phản ánh sự phổ biến của chúng trong thị trường âm nhạc.
2. Thể loại nhạc Techno có độ phổ biến trung bình thấp nhất (37.100000). Điều này có thể cho thấy rằng Techno không được nhiều người nghe.
3. Các thể loại nhạc khác nhau có mức độ phổ biến rất đa dạng, từ rất phổ biến như Pop và Rap đến ít phổ biến như Techno và Psytrance.

* Question 2: So sánh các chỉ số như "speechiness", "acousticness", "instrumentalness", và "liveness" giữa các bài hát trong playlist "Pop Remix" và "Dance Pop".

```python
top_playlists = data_merged['playlist_name'].value_counts().nlargest(4).index.tolist()
print(top_playlists)
```

```python
# Lọc dữ liệu để chỉ bao gồm các bài hát trong hai playlist phổ biến nhất
data_filtered = data_merged[data_merged['playlist_name'].isin(['Trap Americana', 'Dance Pop'])]

# Chọn các cột có kiểu dữ liệu số để tính toán giá trị trung bình
numeric_columns = data_filtered.select_dtypes(include=['float64', 'int64']).columns

# Tính giá trị trung bình của các thuộc tính cho mỗi playlist
data_means = data_filtered.groupby('playlist_name')[numeric_columns].mean().reset_index()

# Chỉ giữ lại các cột cần thiết cho biểu đồ
columns_to_plot = ['playlist_name', 'speechiness_x', 'acousticness_x', 'instrumentalness_x', 'valence_x']
data_means = data_means[columns_to_plot]
print(data_means)
# Tạo biểu đồ đường để so sánh các giá trị trung bình
plt.figure(figsize=(10, 6))

# Vẽ biểu đồ cho từng thuộc tính
plt.plot(data_means['playlist_name'], data_means['speechiness_x'], marker='o', label='Speechiness')
plt.plot(data_means['playlist_name'], data_means['acousticness_x'], marker='o', label='Acousticness')
plt.plot(data_means['playlist_name'], data_means['instrumentalness_x'], marker='o', label='Instrumentalness')
plt.plot(data_means['playlist_name'], data_means['valence_x'], marker='o', label='Valence')

# Thêm tiêu đề và nhãn
plt.title('Comparison of Speechiness, Acousticness, Instrumentalness, and Liveness')
plt.xlabel('Playlist Name')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```

#### Observations:

1. Speechiness: "Trap Americana" có nhiều lời nói hơn "Dance Pop".
2. Acousticness: "Dance Pop" sử dụng âm thanh acoustic nhiều hơn "Trap Americana".
3. Instrumentalness: "Dance Pop" có nhiều nhạc cụ hơn so với "Trap Americana"
4. Valence: "Dance Pop" có cảm giác tích cực, vui vẻ hơn "Trap Americana".

* Question 3: Thời lượng trung bình của các bài hát là bao nhiêu ?

```python
# Tính thời lượng trung bình của các bài hát trong mỗi thể loại nhạc
time_avg = data_merged.groupby('genre')['duration_ms_y'].mean().reset_index()
```

```python
# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
sb.barplot(x='genre', y='duration_ms_y', data=time_avg, hue='genre', dodge=False, palette='viridis', legend=False)
plt.title('Thời lượng trung bình của các thể loại nhạc')
plt.xlabel('Thể loại nhạc')
plt.ylabel('Thời lượng trung bình (ms)')
plt.xticks(rotation=45, ha='right')  # Để tránh chồng lên nhau
plt.tight_layout()
plt.show()
```

#### Observations:

1. Techno có thời lượng trung bình cao nhất
2. Trap Metal có thời lượng trung bình thấp nhất
3. Các thể loại nhạc khác như hiphop, pop, và R&B có thời lượng trung bình khá tương đồng
